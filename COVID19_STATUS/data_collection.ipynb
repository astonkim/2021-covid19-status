## 코로나19 감염현황 - 데이터 수집 (공공데이터 활용)

### Import Modules

```python
# 필요한 모듈을 import 합니다.

import json            # Python에서 JSON을 사용하기 위한 모듈
import requests        # HTTP Requests를 위한 모듈
import xmltodict       # XML 형식을 JSON으로 변환하기 위한 모듈
import pandas as pd    # DataFrame을 생성하고 가공하기 위한 모듈
```

### Create URL for Requests

```python
# Requests를 위한 URL을 생성합니다. 서비스키에 본인의 공공데이터 서비스 키를 입력합니다.

url = "http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?"

url += "serviceKey=8z0vjJIRM7OAedUyFhzT5kc6Lz1VqJB1FO%2FE38W5WxpNiEhY3aG3KTCr6exNt7O82u8wGSUKYOgPvNA1fd2%2Fnw%3D%3D"
url += "&startCreateDt=20210101"
url += "&endCreateDt=20211231"

res = requests.get(url)
res
```

### Inquire res info

```python
# 응답 결과의 내용을 조회합니다. (응답 변수의 text 속성을 조회)

res.text
```

### Convert the form of res info

```python
# Convert res info to the form of Python Dict / 응답 결과의 내용을 Python Dict 형태로 변환합니다.

res_json = xmltodict.parse(res.text)  # 문자열을 XML을 JSON 형식으로 변환합니다.
res_dict = json.dumps(res_json)      # 변환된 JSON 형식을 Dictionary 문자열로 변환합니다.
data = json.loads(res_dict)      # Dictionary 문자열을 파이썬의 Dictionary 형식으로 변환합니다.

data = json.loads(json.dumps(xmltodict.parse(res.text)))  # 1줄로 작성된 변환 과정
```

```python
# Select the needed data / 필요한 데이터만 선택합니다.

data["response"]["body"]["items"]
```

```python
# Convert the selected data to DataFrame / 선택된 데이터를 DataFrame으로 변환합니다.

df = pd.DataFrame(data["response"]["body"]["items"]["item"])
df
```

### Save the result

```python
# seq 컬럼의 값을 기준으로 DataFrame을 오름차순 정렬 후 파일로 저장합니다.

df.sort_values("seq").to_excel("코로나 19 감염 현황.xlsx", index=False)
```

```python
# Read the data / 데이터를 읽어옵니다.

df = pd.read_excel("코로나 19 감염 현황.xlsx")
df
```
