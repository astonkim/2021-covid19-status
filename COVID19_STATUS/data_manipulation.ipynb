# COVID19 2021 - Data Collection
## 코로나19 감염현황 - 데이터 수집 (공공데이터 활용)

### Import Modules

```python
# 필요한 모듈을 import 합니다.

import pandas as pd    # DataFrame을 생성하고 가공하기 위한 모듈
```

### Define the name of the columns

```python
# columns 속성을 이용한 컬럼명 변경 (전체 변경)

df = pd.read_excel("코로나 19 감염 현황.xlsx")
df.colums = ["누적 검사 수", "등록일시분초", "사망자 수", "확진자 수", "게시글번호", "기준일", "기준시간", "수정일시분초", "누적 확진률"]
```

```python
# rename 함수를 이용한 컬럼명 변경 (특정 대상 변경)

df.rename(columns={ "accExamCnt": "누적 검사 수",
                    "createDt": "등록일시분초",
                    "deathCnt": "사망자 수",
                    "decideCnt": "확진자 수",
                    "seq": "게시글번호",
                    "stateDt": "기준일",
                    "stateTime": "기준시간",
                    "updateDt": "수정일시분초",
                    "accDefRate": "누적 확진률" }, inplace=True)
df
```

### Missing Value Imputation

```python
# 결측치 처리

df.fillna(0, inplace=True)
df
```

### Rearrange the data

```python
# 데이터 재정렬

df.sort_values("게시글번호", inplace=True)
df
```

### Remove unnecessary columns

```python
# 불필요한 컬럼 제거

df.drop(["등록일시분초", "수정일시분초", "기준시간", "게시글번호"], axis=1, inplace=True)
df
```

### Convert the data type

```python
# 데이터형 변환

df["기준일"] = pd.to_datetime(df["기준일"], format="%Y%m%d")
df
```

### Remove the duplicate data

```python
# 동일 날짜에 여러 데이터가 존재하는지 여부 확인

df["기준일"].value_counts()
```

```python
# for loop를 이용하여 동일 날짜에 여러 데이터가 존재하는지 여부 확인

for value in df["기준일"].value_counts():
    if value == 1:
        pass
    else:
        print(value)
```

```python
# 만약 중복 데이터가 존재한다면?

"""
만약, 2020-02-08 (value=5) 이라면 ...

df.loc[df["기준일"] == "2020-02-08", :] # 중복데이터의 값을 확인하고 싶을 때, loc 인덱서로 해당 날짜의 데이터만 조회 가능
df.drop_duplicates("기준일", keep="last", inplace=True) # 중복값을 제거하고 가장 마지막에 수정된 값만 남기기
df
"""
```

### Organize DataFrame 

```python
# 추출된 태그명의 알파벳 순서대로 정렬되어있는 데이터프레임을 보기 편하게 재설정 & 인덱스도 초기화

df = df[["누적 검사 수", "사망자 수", "확진자 수", "기준일", "누적 확진률"]].reset_index(drop=True)
df
```

### Calculate the increase number of deaths and confirmed cases compare to previous day

```python
# 전일 계산

new_columns = ["전일대비 " + name for name in df.columns[1:3]]
df.loc[:, new_columns] = df.iloc[:, 1:3].diff().values
df
```

### Extract date information

```python
# 날짜 정보 추출

df['기준일_연도'] = df['기준일'].dt.year
df['기준일_월'] = df['기준일'].dt.month
df['기준일_요일'] = df['기준일'].dt.day_name()
df['기준일_분기'] = df['기준일'].dt.quarter
df
```

### Save the final processed data

```python
df.to_excel('코로나 19 감염 현황 가공 완료.xlsx')
```
